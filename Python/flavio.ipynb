# Introducción a *flavio*

El paquete *flavio* sirve para calcular los valores de un gran número de observables físicos, tanto en el SM como usando teorías efectivas. La mayor parte de los observables están relacionados con la física del mesón \(B\), pero también observables electrodébiles, de Higgs, violación del sabor,...

```python
import flavio
```

Podemos ver la lista de todos los inputs que usa *flavio* para calcular los observables usando el sigiente comando

```python
flavio.default_parameters.all_parameters
```

Los primeros inputs corresponden con parámetros del Lagrangiano del SM: La constante de Fermi `GF`, las constantes de acoplo `alpha_e` y `alpha_s`, el seno cuadrado del ángulo de Weinberg `s2w` y los elementos de la matriz CKM `Vus`, `Vub`, `Vcb` y `delta` (fase CP), el resto de elementos de la CKM se determinan por unitariedad. Los demás inputs son factores de forma hadrónicos.

Para cada input, *flavio* contiene su valor central y su error experimental

```python
flavio.default_parameters.get_central('GF')
```

```python
flavio.default_parameters.get_1d_errors_rightleft()['GF']
```
